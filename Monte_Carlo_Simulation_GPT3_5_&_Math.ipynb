```python
def solve_math(problem: str, temperature: float=0) -> float:
  llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=temperature)
  llm_math = LLMMathChain.from_llm(llm, verbose=True)

  res = llm_math.run(problem)

  matches = re.findall("(?<=Answer: ).*", res)

  return float(matches[0].strip())
```

```python
def generate_problem(n_operands:int=2) -> str:
  if n_operands < 2:
    raise Exception('n_operands must be at least 2')

  operand_set = range(-1000, 1000)
  operator_set = ['+', '-', '*', '/']

  operands = random.choices(operand_set, k=n_operands)
  operators = random.choices(operator_set, k=n_operands-1)

  problem = ''
  for i, o in enumerate(operands):
    problem = problem + f'{o}'
    # print(problem)
    if i < len(operators):
      problem = problem + f'{operators[i]}'
      # print(problem)

  return problem
```

```python
%%time
ans_list = []
for i in range(100):
  print(i)
  time.sleep(61)
  problem = generate_problem(random.randint(2, 5))
  py_ans = float(eval(problem))
  for t in [0, 0.5, 1]:
    ans = solve_math(problem, temperature=t)
    ans_list.append({
        'id': i+1,
        'problem': problem,
        'py_ans': py_ans,
        'temperature': t,
        'llm_ans': ans
    })
```

```python
df = pd.DataFrame(ans_list)
```

```python
df['diff'] = df['py_ans'] - df['llm_ans']
```
